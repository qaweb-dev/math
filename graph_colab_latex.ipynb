# Приклад 2. Зірка в середині кола

```python
%%tikz -S star_in_circle.tikz -f svg
\draw (0,0) circle (0.7);
\draw[fill=yellow] (90:0.5) -- (162:0.5) -- (234:0.5) -- (306:0.5) -- (18:0.5) -- cycle;
```

# Приклад 3. Сітка квадратів

```python
%%tikz -S grid_of_squares.tikz -f svg
\foreach \x in {0,1,2,3} {
    \foreach \y in {0,1,2,3} {
        \draw (\x,\y) rectangle (\x+1,\y+1);
    }
}
```

# Приклад 5. Дуги кола різного радіуса

```python
%%tikz -S circular_arcs.tikz -f svg
\draw (0,0) circle (1);
\draw[red] (0:1) arc (0:90:1);
\draw[blue] (0:0.5) arc (0:90:0.5);
\draw[green] (90:1) arc (90:180:1);
```

# Приклад 6. Квітка з пелюстками

```python
%%tikz -S flower_petals.tikz -f svg
\foreach \angle in {0,60,120,180,240,300} {
    \filldraw[fill=purple!30] (\angle:0.5) circle (0.3);
}
\filldraw[fill=yellow] (0,0) circle (0.2);
```

# Приклад 7. Граф з вершинами ребер

```python
%%tikz -S graph.tikz -f svg
\foreach \i in {0,1,2,3,4} {
    \node[circle,fill=blue!50] (\i) at (\i*72:1) {};
}
\foreach \i in {0,1,2,3,4} {
    \foreach \j in {0,1,2,3,4} {
        \ifnum\i<\j
            \draw (\i) -- (\j);
        \fi
    }
}
```

# Приклад 8. Хвилясті лінії

```python
%%tikz -S sine_wave.tikz -f svg
\draw[blue] plot[domain=0:4*pi,samples=200] (\x,{sin(\x r)});
\draw[red] plot[domain=0:4*pi,samples=200] (\x,{cos(\x r)});
```

# Приклад 9. Радіальна сітка

```python
%%tikz -S radial_grid.tikz -f svg
\foreach \r in {0.5,1,1.5,2} {
    \draw (0,0) circle (\r);
}
\foreach \angle in {0,30,...,360} {
    \draw (0,0) -- (\angle:2);
}
```

# Приклад 10. Складна квітка

```python
%%tikz -S complex_flower.tikz -f svg
\foreach \i in {0,1,...,11} {
    \draw[rotate around={\i*30:(0,0)},red] (0,0) ellipse (1.5cm and 0.5cm);
}
\filldraw[fill=yellow] (0,0) circle (0.2);
```

# Приклад 11. Кубічна решітка

```python
%%tikz -S cubic_grid.tikz -f svg
\foreach \x in {0,1,2} {
    \foreach \y in {0,1,2} {
        \foreach \z in {0,1,2} {
            \draw[fill=blue] (\x,\y,\z) circle (0.05);
        }
    }
}
\draw (0,0,0) -- (2,0,0);
\draw (0,0,0) -- (0,2,0);
\draw (0,0,0) -- (0,0,2);
```

# Приклад 12. 3D параболоїд

```python
%%tikz -S paraboloid.tikz -f svg
\foreach \r in {0,0.2,...,2} {
    \draw[blue] plot[domain=0:360,samples=100] ({\r*cos(\x)}, {\r*sin(\x)}, {\r*\r});
}
\foreach \angle in {0,45,...,360} {
    \draw[red] plot[domain=0:2,samples=100] ({\x*cos(\angle)}, {\x*sin(\angle)}, {\x*\x});
}
```

# Приклад 13. Епітрохоїда

```python
%%tikz -S epitrochoid.tikz -f svg
\draw[blue] plot[domain=0:6.28,samples=200]
    ({cos(\x r) + 0.5*cos(5*\x r)}, {sin(\x r) + 0.5*sin(5*\x r)});
```

# Приклад 14. Тривимірна синусоїдна поверхня

```python
%%tikz -S sine_surface.tikz -f svg
\foreach \x in {-3,-2.5,...,3} {
    \foreach \y in {-3,-2.5,...,3} {
        \draw[blue] (\x,\y,{sin(deg(sqrt(\x*\x+\y*\y)))}) --
                    (\x+0.5,\y,{sin(deg(sqrt((\x+0.5)*(\x+0.5)+\y*\y)))});
        \draw[blue] (\x,\y,{sin(deg(sqrt(\x*\x+\y*\y)))}) --
                    (\x,\y+0.5,{sin(deg(sqrt(\x*\x+(\y+0.5)*(\y+0.5))))});
    }
}
```

# Приклад 15. Розетка Лісажу

```python
%%tikz -S lissajous_rose.tikz -f svg
\foreach \a in {1,2,3,4} {
    \draw[blue] plot[domain=0:2*pi,samples=200]
        ({sin(\a*\x r)}, {cos((\a+1)*\x r)});
}
```

#Приклад 16. Гіперболічна поверхня

```python
%%tikz -S hyperbolic_surface.tikz -f svg
\foreach \x in {-2,-1.8,...,2} {
    \foreach \y in {-2,-1.8,...,2} {
        \draw[red] (\x,\y,{\x*\y}) -- (\x+0.2,\y,{\x*\y+0.2*\y});
        \draw[red] (\x,\y,{\x*\y}) -- (\x,\y+0.2,{\x*\y+0.2*\x});
    }
}
```
